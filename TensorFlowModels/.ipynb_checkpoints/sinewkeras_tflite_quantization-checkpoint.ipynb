```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2' 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

import tensorflow_model_optimization as tfmot
```

```python

import tempfile
import pathlib

tmpdir = tempfile.mkdtemp()
```

```python

x=np.linspace(0,2*np.pi,360)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python

#model = Sequential([Dense(10, input_shape=(1,)), Activation('tanh'), Dense(3),Activation('tanh'),Dense(1)])
model = keras.Sequential([layers.Dense(units=4, input_shape=[1])])
model.add(layers.Dense(units=10,activation='tanh')) #hidden layer
model.add(layers.Dense(units=10,activation='tanh')) #hidden layer
model.add(layers.Dense(units=10,activation='tanh')) #hidden layer
model.add(layers.Dense(units=10,activation='tanh')) #hidden layer


model.add(layers.Dense(units=1))
```

```python
#opt = keras.optimizers.Adam(learning_rate=0.0001)

model.compile(loss='mean_squared_error', optimizer='SGD', metrics=['mean_squared_error'])

```

```python
history = model.fit(x,y,batch_size=len(x),  epochs=5000)
```

```python
plt.plot(history.history['mean_squared_error'], label='mean_squared_error')
plt.figure()
plt.plot(history.history['loss'], label='loss')
#plt.plot(history.history['val_loss'], label='val_loss')
#plt.ylim([0, 10])
plt.xlabel('Epoch')
```

```python
predictions = model.predict(x)

plt.plot(x, y, 'b', x, predictions, 'r--')
plt.plot(x, predictions, 'r--')

#plt.ylabel('Y / Predicted Value')
#plt.xlabel('X Value')
```

```python
net_save_path = os.path.join(tmpdir, "net/1/")
savedir="/home/fred/"
net_save_path = os.path.join(tmpdir, "net/1/")
tf.saved_model.save(model, net_save_path)
```
