```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2' 
import tensorflow as tf
import numpy as np
from tensorflow import keras

import tempfile
import pathlib

tmpdir = tempfile.mkdtemp()
```

```python
MODELS_DIR = 'models/'
if not os.path.exists(MODELS_DIR):
    os.mkdir(MODELS_DIR)
MODEL_TF = MODELS_DIR + 'model'
MODEL_NO_QUANT_TFLITE = MODELS_DIR + 'model_no_quant.tflite'
MODEL_TFLITE = MODELS_DIR + 'model.tflite'
MODEL_TFLITE_MICRO = MODELS_DIR + 'model.cc'
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=4, input_shape=[2,1])])
model.add(keras.layers.Dense(units=4)) #hidden layer 2
model.add(keras.layers.Dense(units=1)) #output layer   
```

```python
xs = np.array([0.0, 0.5, 1.0], dtype=float)
ys = np.array([1.0, -1.0, 1], dtype=float)
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(xs, ys, epochs=500)
```

```python
net_save_path = os.path.join(tmpdir, "net/1/")
savedir="/home/fred/"
net_save_path = os.path.join(tmpdir, "net/1/")
tf.saved_model.save(model, net_save_path)
```

```python
converter = tf.lite.TFLiteConverter.from_saved_model(net_save_path) # path to the SavedModel directory
tflite_model = converter.convert()

# Save the model.
with open('models/model.tflite', 'wb') as f:
  f.write(tflite_model)
```
